```python
df = pd.read_parquet(os.path.join(DATA_PATH, "intermediate_data/cpc", "analysis_data_w_distances_suffixes.parquet"))
cluster_centroids = pd.read_parquet(os.path.join(DATA_PATH, "intermediate_data/cpc", "cluster_centroids.parquet"))
embeddings_pca = pd.read_parquet(os.path.join(DATA_PATH, "intermediate_data/cpc", "embeddings_pca.parquet"))
```

```python
df.to_csv(os.path.join(DATA_PATH, "intermediate_data/cpc", "analysis_data_w_distances_suffixes.csv"))

centroids_out = []
for idx, row in cluster_centroids.iterrows():
    outrow = {'cluster': row['cluster']}
    for i in range(len(row['embedding_pca'])):
        outrow[f'd{i}'] = row['embedding_pca'][i]
    centroids_out.append(outrow)
centroids_out_df = pd.DataFrame(centroids_out)
centroids_out_df.to_csv(os.path.join(DATA_PATH, "intermediate_data/cpc", "cluster_centroids.csv"), index=False)

embeddings_pca_out = []
for idx, row in embeddings_pca.iterrows():
    outrow = {
        'date': row['date'],
        'year': row['year'],
        'item_no': row['item_no'],
        'cluster': row['cluster']
    }
    for i in range(len(row['embedding_pca'])):
        outrow[f'd{i}'] = row['embedding_pca'][i]
    embeddings_pca_out.append(outrow)
embeddings_pca_out_df = pd.DataFrame(embeddings_pca_out)
embeddings_pca_out_df.to_csv(os.path.join(DATA_PATH, "intermediate_data/cpc", "embeddings_pca.csv"), index=False)

```
